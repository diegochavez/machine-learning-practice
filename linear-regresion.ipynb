```python
dummies = pd.get_dummies(clean_data["ocean_proximity"], dtype=int)
clean_data = pd.concat([clean_data, dummies], axis=1)
clean_data = clean_data.drop("ocean_proximity", axis=1)
clean_data.head()
```

```python
X = clean_data.drop("median_house_value", axis=1)
```

```python
Y = clean_data["median_house_value"]
```

```python
# let's split the data into training and testing sets
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()

model.fit(X_train, Y_train)
```

```python
predictions = model.predict(X_test)
```

```python
pd.DataFrame({ "Predicted": predictions, "Actual": Y_test,})
```

```python
# Overfitting check
print(model.score(X_train, Y_train))
print(model.score(X_test, Y_test))
```

```python
# Error check
from sklearn.metrics import mean_squared_error
import numpy as np

mse = mean_squared_error(Y_test, predictions)
```

```python
rmse = np.sqrt(mse)
```
