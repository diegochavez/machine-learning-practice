```python
clean_data = data.dropna()
```

```python
clean_data["ocean_proximity"].value_counts()
```

```python
pd.get_dummies(clean_data["ocean_proximity"], dtype=int)
```

```python
dummies = pd.get_dummies(clean_data["ocean_proximity"], dtype=int)
clean_data = pd.concat([clean_data, dummies], axis=1)
clean_data = clean_data.drop("ocean_proximity", axis=1)
clean_data.head()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(25, 10))
sb.heatmap(clean_data.corr(), annot=True, cmap='YlGnBu')
plt.show()
```
