```python
import pandas as pd
import seaborn as sb

datax = pd.read_csv("height_weights.csv")
datax.head()

```

```python
sb.scatterplot(x="height", y="weight", data=datax)
```

```python
# Characteristic of the data
x = datax["height"]
y = datax["weight"]

print(x)
```

```python
x_processed = x.values.reshape(-1,1)
y_processed = y.values.reshape(-1,1)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
# training the model
model.fit(x_processed, y_processed)
```

```python
height = 1.85
model.predict([[height]])
```
